### 1. Different countries were the most successful in summer and winter games?

#### Querrying data from tables

```python
top10_overall=pysqldf("""
        SELECT
            noc,
            COUNT(*)
        FROM medals
        GROUP BY noc
        ORDER BY Count(*) DESC
        LIMIT 10
        """)
```

```python
summer_top10=pysqldf("""
        SELECT
            noc,
            COUNT(*) as medals,
            season
        FROM medals
            JOIN games
            ON games.start_id=medals.start_id
        WHERE season='Summer'
        GROUP BY season, noc
        ORDER BY Count(*) DESC
        LIMIT 10
        """)
```

```python
winter_top10=pysqldf("""
        SELECT
            noc,
            COUNT(*) as medals,
            season
        FROM medals
            JOIN games
            ON games.start_id=medals.start_id
        WHERE season='Winter'
        GROUP BY season, noc
        ORDER BY Count(*) DESC
        LIMIT 10
        """)
```

#### Visualising data using seaborn:

```python
#setting plot
fig, axes = plt.subplots(1, 3, figsize=(30,10))
sns.barplot(data= top10_overall, x= 'NOC', y='COUNT(*)', ax=axes[0], color= 'gray')
sns.barplot(data= summer_top10, x= 'NOC', y='medals', ax=axes[1], color= 'orange')
sns.barplot(data= winter_top10, x= 'NOC', y='medals', ax=axes[2], color= 'cyan')

#setting labels
axes[0].set_ylabel('medals')
axes[0].set_xlabel('country')
axes[1].set_xlabel('country')
axes[2].set_xlabel('country')
axes[0].set_xlabel('country')

#setting titles
axes[0].set_title('Top 10 countries by medals overall', fontsize= 18)
axes[1].set_title('Top 10 countries by medals (summer)', fontsize= 18)
axes[2].set_title('Top 10 countries by medals (winter)', fontsize= 18)

# Unify y-axes
max_y = max(top10_overall['COUNT(*)'].max(), summer_top10['medals'].max(), winter_top10['medals'].max())
for ax in axes:
    ax.set_ylim(0, max_y)
```

Generaly the countries which were the most successful in winter are the same as the countries which were the most successful in summer. Countries which were most sucessful overall recieved more medals durring summer Olympics. Although the order changes depending on season most of the countries apear the same.

Findings:
* China was not in top 10 countries overall but was in top 10 of Summer Olympics.
* In Winter Olympics there were Countries such as Finland, Norway, Sweden, Switzerlnad, Austria.
* Australia, China and France were not among top 10 countries of Winter Olympics.

## Investigating height and weight characteristics of various discipline athletes:
### Checking average height and weight of Basketball players.

```python
pysqldf("""
        SELECT
            AVG(weight),
            AVG(height)
        FROM athlete
            JOIN sport
            ON athlete.start_id=sport.start_id
            INNER JOIN medals
            ON athlete.start_id=medals.start_id
        WHERE sport='Basketball'
        """)
```

### Additional question: Medalist Basketball male players are higher than ones with no medals

```python
pysqldf("""
        SELECT
            AVG(weight),
            AVG(height),
            medal
        FROM athlete
            JOIN sport
            ON athlete.start_id=sport.start_id
            JOIN medals
            ON athlete.start_id=medals.start_id
        WHERE sport='Basketball'
                AND sport.sex='M'
        GROUP BY medal
        """)
```

```python
pysqldf("""
        SELECT
            AVG(weight),
            AVG(height),
            medal
        FROM athlete
            JOIN sport
            ON athlete.start_id=sport.start_id
            LEFT JOIN medals
            ON athlete.start_id=medals.start_id
        WHERE sport='Basketball'
                AND sport.sex='M'
                AND medal IS NULL
        GROUP BY medal
        """)
```

### Joining 2 above queries with "UNION".

```python
basketball_players = pysqldf("""
        SELECT
            AVG(height) AS avg_height,
            medal
        FROM athlete
            JOIN sport
            ON athlete.start_id=sport.start_id
            JOIN medals
            ON athlete.start_id=medals.start_id
        WHERE sport='Basketball'
                AND sport.sex='M'
        GROUP BY medal
        
        UNION
        
        SELECT
            AVG(height) AS avg_height,
            medal
        FROM athlete
            JOIN sport
            ON athlete.start_id=sport.start_id
            LEFT JOIN medals
            ON athlete.start_id=medals.start_id
        WHERE sport='Basketball'
                AND sport.sex='M'
                AND medal IS NULL
        GROUP BY medal
        
        ORDER BY AVG(height) DESC
        
        """)
basketball_players
```

Basketball medalists on average were arround 3 cm lower. This is however should be teste for statistical significance to ensure that (as it is not the objective of this notebook it will not be done here. The process would be very simmilar to the one displayed in my A/B testing notebook.)

## Feature engineeering: New metric created BMI

```python
BMI_medal= pysqldf("""
        SELECT
            id,
            athlete.name,
            weight/(height*height)*10000 AS BMI,
            medal,
            (CASE
                WHEN medal='Gold' THEN 3
                WHEN medal='Silver' THEN 2
                WHEN medal='Bronze' THEN 1
                ELSE 0
            END) AS points
            
        FROM
            athlete
        JOIN medals
           ON athlete.start_id=medals.start_id
        
        UNION
        
        SELECT
            id,
            athlete.name,
            weight/(height*height)*10000 AS BMI,
            medal,
            (CASE
                WHEN medal='Gold' THEN 3
                WHEN medal='Silver' THEN 2
                WHEN medal='Bronze' THEN 1
                ELSE 0
            END) AS points
        FROM
            athlete
        LEFT JOIN medals
           ON athlete.start_id=medals.start_id
        WHERE medal IS NULL
        ORDER BY BMI
        """)

BMI_medal.head(10)

np.corrcoef(BMI_medal['BMI'], BMI_medal['points'])
```

There is no correlation between `BMI` and `medal_points`.
